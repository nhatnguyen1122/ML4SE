```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('dataset_comp_adjust.csv')

plt.figure(figsize=(8, 6))
plt.scatter(data['ratio_0_1'], data['AUC'], color='blue')
plt.xlabel('ratio_0_1')
plt.ylabel('AUC')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(data['dataset_size'], data['AUC'], color='red')
plt.xlabel('dataset_size')
plt.ylabel('AUC')
plt.grid(True)
plt.show()
```
