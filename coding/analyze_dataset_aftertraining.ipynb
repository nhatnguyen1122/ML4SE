```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('dataset_comparison.csv')

df
```

```python
corr = df[['class_0_proportion', 'class_1_proportion', 'dataset_size', 'AUC']].corr()

corr
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', vmin=-1, vmax=1)
plt.title("Correlation Matrix")
plt.show()
```

```python
# Scatter plot between dataset size and AUC
plt.figure(figsize=(8, 6))
sns.scatterplot(x='dataset_size', y='AUC', data=df)
plt.title("Dataset Size vs AUC")
plt.xlabel("Dataset Size")
plt.ylabel("AUC Score")
plt.show()
```

```python
# Pairplot for dataset features
sns.pairplot(df[['class_0_proportion', 'class_1_proportion', 'dataset_size', 'AUC']])
plt.show()
```

```python
df['0_1_ratio'] = df['class_0_proportion'] / df['class_1_proportion']
```

```python
# Scatter plot for 0/1 ratio vs AUC
plt.figure(figsize=(8, 6))
sns.scatterplot(x='0_1_ratio', y='AUC', data=df)
plt.title("0/1 Ratio vs AUC")
plt.xlabel("0/1 Ratio (False Alarms / Actionable Alarms)")
plt.ylabel("AUC Score")
plt.show()
```
